```python
# convert dataframe to numpy array
df_con = df1['Conservation_score']
df_ran_con = df_ran1['Conservation_score']

con_PUM = df_con.as_matrix()
con_Ran = df_ran_con.as_matrix()
```

```python

Conserve_PUM = np.array([])
Conserve_ran = np.array([])

N = np.count_nonzero(con_PUM)
M = np.count_nonzero(con_Ran)

for i in range(N):
    if type(con_PUM[i]) == float:                    # 3'UTR contain only 1 motif
        Conserve_PUM = np.append(Conserve_PUM, con_PUM[i])
    elif type(con_PUM[i]) != float:                 # 3'UTR contains multiple motifs
        conP1 = con_PUM[i].split("|")
        conP2 = [float(j) for j in conP1]
        for j in range(len(conP2)):
            Conserve_PUM = np.append(Conserve_PUM, conP2)
    else:
        pass

for i in range(M):
    conR1 = con_Ran[i].split("|")
    conR2 = [float(j) for j in conR1]
    for j in range(len(conR2)):
        Conserve_ran = np.append(Conserve_ran, conR2)
```

```python
plt.hist(Conserve_PUM, bins = 200, alpha = 0.8, color = 'c', normed = True, label = 'PUM_motif')
plt.hist(Conserve_ran, bins = 200, alpha = 0.5, color = 'k', normed = True, label = 'Random sequence')

plt.legend()
plt.xlabel("Conservation_score (phyloP)")
plt.ylabel("Frequency")
plt.show()
```

```python
Count = len(df3)

dic = {}

gene = df3.iloc[0,0]
num = df3.iloc[0,1]
seq = df3.iloc[0,2].split('|')
pos = df3.iloc[0,3].split('|')
length = df3.iloc[0,4]
con = df3.iloc[0,3].split('|')

for i in range (int(num)):
    gene_name = gene +'_'+ str(i)
    dic[gene_name] = [seq[i], int(pos[i]), int(length), int(con[i])]
```

```python
Count = len(df3)

dic = {}

for j in range (int(Count)):
    
    gene = df3.iloc[j,0]
    num = df3.iloc[j,1]
    if num == 1:
        seq = df3.iloc[j,2]
        pos = df3.iloc[j,3]
        length = df3.iloc[j,4]
        con = df3.iloc[j,5]
        gene_name = gene
        dic[gene_name] = [seq, int(pos), int(length), float(con)]
    elif num >1:
        seq = df3.iloc[j,2].split('|')
        pos = df3.iloc[j,3].split('|')
        length = df3.iloc[j,4]
        con = df3.iloc[j,5].split('|')

        for i in range (int(num)):
            gene_name = str(gene) +'_'+ str(i)
            dic[gene_name] = [seq[i], int(pos[i]), int(length), float(con[i])]
    else:
        pass
```

```python
df_sep = pandas.DataFrame.from_dict(dic, orient = 'index')
df_sep.columns = ['PUM motif', 'position', 'UTR length', 'phyloP']

df_sep['distance'] = df_sep['UTR length'] - df_sep['position']

df_sep.sort_index()
```

```python
df_sep_high = df_sep[df_sep['phyloP'] >1]
df_sep_low = df_sep[df_sep['phyloP'] <1]
```

```python
plt.hist(df_sep_high['distance'], bins = 3000, color = 'r', alpha = 0.5, label = 'Conserved PUM motif')
plt.hist(df_sep_low['distance'], bins = 1000, color = 'gray', alpha = 0.5, label = 'unconserved PUM motif')
plt.xlabel('Distance from 3end (bp)')
plt.ylabel('Number of PUM motif')
plt.legend()
plt.xlim(0, 1000)
plt.show()
```
