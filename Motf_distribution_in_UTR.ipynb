```python
# Ratio == calculate the distribution of PUM motif in 3'UTR 
# Diff == calculate distance between 3'end and the motif position

diff = np.array([])
ratio = np.array([])

N = np.count_nonzero(pos[:, 1])

for i in range(N):
    if type(pos[i, 0]) == float:
        diff_i = pos[i, 1] - pos[i, 0]
        diff= np.append(diff, diff_i)
        if pos[i, 1] != 0:
            ratio_i = pos[i, 0]/pos[i, 1]
            ratio = np.append(ratio, ratio_i)
        else:
            pass
    elif type(pos[i, 0]) != float:
        pos1 = pos[i, 0].split("|")
        pos2 = [float(j) for j in pos1]
        for j in range(len(pos2)):
            diff_i = pos[i, 1] - pos2[j]
            ratio_i = pos2[j]/pos[i, 1]
            diff = np.append(diff, diff_i)
            ratio = np.append(ratio, ratio_i)
    else:
        pass
Diff = diff[np.logical_not(np.isnan(diff))]
Ratio = ratio[np.logical_not(np.isnan(ratio))]
```

```python
plt.hist(Diff, bins = 9000, alpha = 0.8)
plt.xlabel("Distance from 3'end (bp)")
plt.ylabel("Number of PUM motif")
plt.show()
```

```python
plt.hist(Ratio, bins = 250, alpha = 0.8)
plt.xlabel("Relative position of PUM motif")
plt.ylabel("Number of PUM motif")
plt.show()
```
